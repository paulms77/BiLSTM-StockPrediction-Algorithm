```python
news_dfs = pd.DataFrame()
news_folder_dir = 'my_news_data/'
for symbol_code in symbol_code_ls:
    print(f'Symbol Code: {symbol_code}')
    file_path = news_folder_dir + f'{symbol_code}_news.csv'
    news_data = pd.read_csv(f'{file_path}')
    news_data.Code = news_data.Code.astype(str).str.zfill(6)
    news_dfs = pd.concat([news_dfs, news_data], axis = 0)
news_dfs.rename(columns = {'Code': 'SYMBOL'}, inplace = True)
news_dfs.rename(mapper=str.upper, axis = 'columns', inplace = True)
```

```python
start = time.time()

# 날짜별 종목별 뉴스 제목들을 하나의 제목으로 연결
grouped_news_dfs = news_dfs.groupby(['DATE', 'SYMBOL'])['TITLE'].apply(' '.join).reset_index()

# Custom Tokenizer 실행
grouped_news_dfs['FILTERED_TITLE'] = grouped_news_dfs['TITLE'].apply(tokenizer_filter)
name_set = set(df_symbols['NAME'])

# 주식 종목명이 뉴스 제목에 포함될 경우 제거
for index, row in grouped_news_dfs.iterrows():
    filtered_title = row['FILTERED_TITLE']
    filtered_title = [word for word in filtered_title if word not in name_set]
    grouped_news_dfs.at[index, 'FILTERED_TITLE'] = filtered_title

filterd_word_set = get_freq_word_set(grouped_news_dfs)

# 날짜별 종목별 정렬
grouped_news_dfs.sort_values(['DATE', 'SYMBOL'], ascending = True)

# 종목들의 일간정보와 뉴스정보를 통합
final_dfs = merge_info_dfs.merge(grouped_news_dfs, on = ['DATE', 'SYMBOL'], how = 'left')

# 뉴스 제목이 없는 경우 결측치로 저장이 되었으므로, 공백 처리
final_dfs['FILTERED_TITLE'] = final_dfs['FILTERED_TITLE'].fillna(' ')

# 감성분석 수행
final_dfs = get_emotional_score(final_dfs, filterd_word_set)

print('감성분석 작업시간: ', time.time() - start)
```

```python
start = time.time()

n_epoch = 20
batch_size = 64
learning_rate = 0.01

n_hidden = 50
n_latent = 10
n_layers = 1

save_dir = 'my_path/xgb'
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
    
save_dir = 'my_path/vae'
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

# feature importance
stacked_symbol_id = []
stacked_vae_filter_data = []
zero_divisions = []
for idx, symbol_id in enumerate(final_dfs['SYMBOL'].unique().tolist()):
    print('-'*20)
    print(f'symbol_id: {symbol_id}')
    print('-'*20)

    final_temp = final_dfs[final_dfs['SYMBOL'] == symbol_id]

    print('Before feature importances', final_temp.shape)

    num_training_days = int(np.ceil(0.8 * final_temp.shape[0]))
    
    if final_temp[num_training_days:].shape[0] == 0:
        print(f'{symbol_id} is Small Data')
        zero_divisions.append(symbol_id)
        continue

    columns_list = final_temp.drop(['SYMBOL'], axis = 1).columns.tolist()

    # scaler = MinMaxScaler()
    # scaler = scaler.fit(final_temp.drop(['SYMBOL', 'CLOSE'], axis = 1))
    # final_temp_scaled = scaler.transform(final_temp.drop(['SYMBOL', 'CLOSE'], axis = 1))
    # print(final_temp_scaled.shape, final_temp['CLOSE'].shape)
    # final_temp_scaled = np.hstack((final_temp_scaled, np.array(final_temp['CLOSE']).reshape(-1, 1)))
    # final_temp_scaled_df = pd.DataFrame(final_temp_scaled, columns = columns_list)
    
    scaler = MinMaxScaler()
    scaler = scaler.fit(final_temp.drop(['SYMBOL'], axis = 1))
    final_temp_scaled = scaler.transform(final_temp.drop(['SYMBOL'], axis = 1))
    final_temp_scaled_df = pd.DataFrame(final_temp_scaled, columns = final_temp.drop(['SYMBOL'], axis = 1).columns.tolist())
    
    # temp = final_dfs[final_dfs['SYMBOL'] == symbol_id]
    # temp = temp.drop(['SYMBOL'], axis = 1)

    (X_train_fi, y_train_fi), (X_test_fi, y_test_fi) = train_test_split(final_temp_scaled_df, 'CLOSE')
    xgbregressor = xgb.XGBRegressor() # gamma=0.0,n_estimators=150,base_score=0.7,colsample_bytree=1,learning_rate=0.05
    xgbregressor_fit = xgbregressor.fit(X_train_fi, y_train_fi, eval_set = [(X_train_fi, y_train_fi), (X_test_fi, y_test_fi)], verbose = False)
    
    # XGBRegressor 학습파일 저장
    xgb_path = f'my_path/xgb/{symbol_id}_xgb.joblib'
    joblib.dump(xgbregressor_fit, xgb_path)

    # XGBRegressor 학습파일 불러오기
    xgbregressor = joblib.load(xgb_path)
    #xgbregressor = joblib.load(os.path.join(os.path.dirname(__file__), xgb_path))
    
    # # feature_importance plot
    # plt.figure(figsize = (len(final_dfs['SYMBOL'].unique().tolist()), len(final_dfs['SYMBOL'].unique().tolist())))
    # feature_importance = xgbregressor.feature_importances_
    # sorted_idx = np.argsort(feature_importance)
    # plt.barh(range(len(sorted_idx)), feature_importance[sorted_idx], align = 'center')
    # plt.yticks(range(len(sorted_idx)), np.array(X_test_fi.columns)[sorted_idx])
    # plt.xlim([0.0, feature_importance.max()])
    # plt.title(f'{symbol_id} Feature Importance')
    # plt.show()
    explainer = shap.Explainer(xgbregressor)
    shap_values = explainer(X_test_fi)
    #shap.plots.waterfall(shap_values[0])
    shap.plots.beeswarm(shap_values)

    #fi = X_train_fi.columns[np.array(xgbregressor.feature_importances_) > 0.0].tolist()
    fi = X_train_fi.columns[np.abs(shap_values[0].values) > 0.0].tolist()
    print('len:', len(fi))
    print(fi)
    fi.append('CLOSE')

    temp = final_dfs[final_dfs['SYMBOL'] == symbol_id]
    temp = temp[list(set(fi))]
    print('after feature importances', temp.shape)

    if temp.shape[1] < 2:
        print(f'{symbol_id} is zero')
        stacked_vae_filter_data.append(torch.Tensor(temp['CLOSE'].values))
        continue

    scaler = MinMaxScaler(feature_range = (0, 1))
    scaler = scaler.fit(temp.drop(['CLOSE'], axis = 1))
    temp_scaled = scaler.transform(temp.drop(['CLOSE'], axis = 1))

    num_training_days = int(np.ceil(0.8 * temp.shape[0]))
        
    feats_temp = temp_scaled[:, :]
    feats_train = temp_scaled[:num_training_days, :]
    feats_test = temp_scaled[num_training_days:, :]
    
    data_close = torch.tensor(temp['CLOSE'].values)

    # Stacked AutoEncoder
    feats_train_tensor = torch.tensor(feats_train, dtype = torch.float32)
    feats_test_tensor = torch.tensor(feats_test, dtype = torch.float32)

    train_loader = DataLoader(feats_train_tensor,  batch_size = batch_size, shuffle = False)
    test_loader = DataLoader(feats_test_tensor, batch_size = batch_size, shuffle = False)

    if feats_train.shape[1] > 64:
        n_hidden = 64
    else:
        n_hidden = 32

    model = train_vae(train_loader, test_loader, feats_train, symbol_id, n_hidden, n_latent, n_layers, learning_rate, n_epoch, True)

    feats_data = torch.tensor(feats_temp, dtype = torch.float32)
    mu, lv = model.encoded(feats_data)
    z = model.reparameterize(mu, lv)
    decoded_data = model.decoded(z)
    print(decoded_data.shape)
    
    decoded_data = torch.cat((torch.tensor(scaler.inverse_transform(decoded_data.detach().numpy())), data_close.view(data_close.shape[0], 1)), dim = 1)
    print(decoded_data.shape)
    stacked_symbol_id.append(symbol_id)
    stacked_vae_filter_data.append(decoded_data)

#stacked_vae_filter_data = {idx:torch.tensor(data) for idx, data in enumerate(stacked_vae_filter_data)}

stacked_vae_filter_data = {symbol_id:torch.tensor(data) for symbol_id, data in zip(stacked_symbol_id, stacked_vae_filter_data)}

print('기능 중요도 및 상위 수준 기능 추출 작업시간: ', time.time() - start)
```

```python
print(torch.cuda.is_available())
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
print(device)

start = time.time()

# BiLSTM
prediction_window = 5
window_size = 120

num_epochs = 500
learning_rate = 0.01
batch_size = 64

hidden_size = 50
n_layers = 1
dropout = 0

model_predictions = {}
returns_dailys = {}
zero_sequences = []

f, axes = plt.subplots(5, 2)
f.set_size_inches((20, 15))
plt.subplots_adjust(wspace = 0.3, hspace = 0.3)
jdx = 0

for i, symbol_id in enumerate(stacked_vae_filter_data.keys()):
    print('-'*20)
    print(f'symbol_id: {symbol_id}')
    print('-'*20)
    temp = stacked_vae_filter_data[symbol_id]
    last_return = temp[-1][-1]
    
    if i == 0:
        idx = 0
    elif i == 1:
        idx = 1
    
    elif i > 1:
        if i % 2 == 0:
            idx = 0
            jdx = i // 2
        else:
            idx = 1
            jdx = i // 2
        
    # emotional_weights = final_dfs.loc[fianl_dfs.SYMBOL == symbol_id, 'EMOTIONAL_SCORES'].iloc[-1].values
    
    symbol_test_stock = fdr.DataReader(symbol_id, '2023-11-11', '2023-11-29')
    symbol_test_close = symbol_test_stock['Close'][:prediction_window]

    sequences = []
    targets = []
    if len(temp.shape) < 2:
        print(f'{symbol_id} is Small Data')
        scaler = MinMaxScaler(feature_range = (0, 1))
        temp = scaler.fit_transform(temp.detach().numpy().reshape(-1, 1))
        temp = torch.tensor(temp, dtype = torch.float32)
        temp = temp.squeeze()

        for i in range(len(temp) - window_size - prediction_window + 1):
            sequences.append(temp[i: i+window_size])
            targets.append(temp[i+window_size: i+window_size+prediction_window])
            
        input_size = 1

        if len(sequences) == 0:
            zero_sequences.append(symbol_id)
            continue
        
        sequences = torch.stack(sequences)
        targets = torch.stack(targets)

        print('sequences shape: ', sequences.shape)
        print('targets shape: ', targets.shape)

        train_dataset = TensorDataset(sequences[:-2], targets[:-2])
        train_loader = DataLoader(train_dataset, batch_size = batch_size, shuffle = False)
        
        output_size = prediction_window
        
        model = train_bilstm(train_loader, symbol_id, input_size, hidden_size, n_layers, dropout, output_size, num_epochs, learning_rate, device, True)
        
        train_outputs = bilstm_inference(sequences[-1].unsqueeze(0), symbol_id, input_size, hidden_size, n_layers, dropout, output_size, window_size, device, True) # sequences[-1].unsqueeze(0)
    
        train_outputs = scaler.inverse_transform(train_outputs.cpu().numpy().reshape(-1, 1))
        
        model_prediction = train_outputs.reshape(-1, output_size)[-1]
        model_prediction = np.array(model_prediction)
        
        # if emotional_weights > 0
        #     model_prediction *= emotional_weights
        
        #mse = mean_squared_error(model_prediction, symbol_test_close.values)
        #rmse = mean_squared_error(model_prediction, symbol_test_close.values, squared = False)
        
        #print('MSE', mse)
        #print('RMSE', rmse)
        
        predict = info_dfs.loc[info_dfs.SYMBOL == symbol_id, 'CLOSE'] + model_prediction
        
        
        axes[jdx, idx].plot(range(0, info_dfs.loc[info_dfs.SYMBOL == symbol_id, 'CLOSE'].shape[0]), info_dfs.loc[info_dfs.SYMBOL == symbol_id, 'CLOSE'], color = 'b')
        axes[jdx, idx].plot(range(info_dfs.loc[info_dfs.SYMBOL == symbol_id, 'CLOSE'].shape[0]-1, info_dfs.loc[info_dfs.SYMBOL == symbol_id, 'CLOSE'].shape[0]+prediction_window-1), model_prediction, color = 'g', label = 'pred')
        axes[jdx, idx].plot(range(info_dfs.loc[info_dfs.SYMBOL == symbol_id, 'CLOSE'].shape[0]-1, info_dfs.loc[info_dfs.SYMBOL == symbol_id, 'CLOSE'].shape[0]+prediction_window-1), symbol_test_close.values, color = 'r', label = 'true')
        axes[jdx, idx].legend()
        axes[jdx, idx].set_title(f'{symbol_id}')

        returns_daily = (model_prediction[-1] - last_return) / last_return
        
        # model_predictions.append(model_prediction)
        model_predictions[symbol_id] = model_prediction
        #returns_dailys.append(returns_daily)
        returns_dailys[symbol_id] = returns_daily

    else:
        feats_temp = temp[:, :-1].detach().numpy()

        # scaling
        feats_temp_scaled = scaler.fit_transform(feats_temp)

        #pca
        pca = PCA(n_components = .8)
        feats_temp_pca = pca.fit_transform(feats_temp_scaled)
        print('pca n_components_', feats_temp_pca.shape[1])

        feats_temp = torch.tensor(feats_temp_pca, dtype = torch.float32)
        #feats_temp = torch.tensor(feats_temp_scaled, dtype = torch.float32)
        temp = torch.tensor(temp, dtype = torch.float32)

        for i in range(len(temp) - window_size - prediction_window + 1):
            sequences.append(feats_temp[i: i+window_size, :])
            targets.append(temp[i+window_size: i+window_size+prediction_window, -1])

        input_size = feats_temp.shape[1]

        if len(sequences) == 0:
            zero_sequences.append(symbol_id)
            continue
        
        sequences = torch.stack(sequences)
        targets = torch.stack(targets)

        scaler = MinMaxScaler(feature_range = (0, 1))
        normalized_targets = scaler.fit_transform(targets.reshape(-1, 1))
        normalized_targets = normalized_targets.reshape(-1, prediction_window)
        normalized_targets = torch.tensor(normalized_targets, dtype = torch.float32)

        print('sequences shape: ', sequences.shape)
        print('targets shape: ', targets.shape)

        train_dataset = TensorDataset(sequences[:-2], normalized_targets[:-2])
        train_loader = DataLoader(train_dataset, batch_size = batch_size, shuffle = False)
        
        output_size = prediction_window
        
        model = train_bilstm(train_loader, symbol_id, input_size, hidden_size, n_layers, dropout, output_size, num_epochs, learning_rate, device, False)
        
        train_outputs = bilstm_inference(sequences[-1].unsqueeze(0), symbol_id, input_size, hidden_size, n_layers, dropout, output_size, window_size, device, False) # sequences[-1].unsqueeze(0)
    
        train_outputs = scaler.inverse_transform(train_outputs.cpu().numpy().reshape(-1, 1))
        
        model_prediction = train_outputs.reshape(-1, output_size)[-1]
        model_prediciton = np.array(model_prediction)
        
        # if emotional_weights > 0
        #     model_prediction *= emotional_weights
        
        #mse = mean_squared_error(model_prediction, symbol_test_close.values)
        #rmse = mean_squared_error(model_prediction, symbol_test_close.values, squared = False)
        
        #print('MSE', mse)
        #print('RMSE', rmse)
        
        green = np.concatenate((info_dfs.loc[info_dfs.SYMBOL == symbol_id, 'CLOSE'].values, model_prediction))
        red = np.concatenate((info_dfs.loc[info_dfs.SYMBOL == symbol_id, 'CLOSE'].values, symbol_test_close.values))
        
        axes[jdx, idx].plot(range(0, green.shape[0]), green, color = 'g', label = 'pred')
        axes[jdx, idx].plot(range(0, red.shape[0]), red, color = 'r', label = 'true')
        axes[jdx, idx].plot(range(0, info_dfs.loc[info_dfs.SYMBOL == symbol_id, 'CLOSE'].shape[0]), info_dfs.loc[info_dfs.SYMBOL == symbol_id, 'CLOSE'], color = 'b')
        axes[jdx, idx].legend()
        axes[jdx, idx].set_title(f'{symbol_id}')

        returns_daily = (model_prediction[-1] - last_return) / last_return

        # model_predictions.append(model_prediction)
        model_predictions[symbol_id] = model_prediction
        #returns_dailys.append(returns_daily)
        returns_dailys[symbol_id] = returns_daily
        
plt.tight_layout()
plt.show()

print('bi-LSTM 작업시간: ', time.time() - start)
```
